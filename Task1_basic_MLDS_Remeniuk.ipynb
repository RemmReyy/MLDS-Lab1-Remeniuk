Завдання 2: Розв’язати квадратне рівняння $ax^2+bx + c = 0$.  Напишіть функцію,яка приймає параметри квадратного рівняння, а повертає одне рішення або декілька.

Наприклад:  


```
solve_quad(a=1, b=-2, c=1) -> 1  
solve_quad(a=1, b=0, c=-1) -> (-1, 1)  
solve_quad(a=0, b=2, c=-1) -> 0.5
```







````python
import math

def solve_quad(a, b, c):
    # If a is 0, it's a linear equation: bx + c = 0
    if a == 0:
        if b == 0:
            return None  # No solution if b is also 0
        return -c / b

    # Calculate discriminant
    discriminant = b**2 - 4*a*c

    # If discriminant is negative, no real roots
    if discriminant < 0:
        return None

    # If discriminant is zero, one root
    if discriminant == 0:
        return -b / (2*a)

    # If discriminant is positive, two roots
    root1 = (-b + math.sqrt(discriminant)) / (2*a)
    root2 = (-b - math.sqrt(discriminant)) / (2*a)
    return (root1, root2)
````

Завдання 4: Реалізація алгоритму [MergeSort](https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC)..

Наприклад: `MergeSort([3, 1, 3, 5]) -> [1, 3, 3, 5]`

````python
def MergeSort(L: List[int]) -> List[int]:
    if len(L) <= 1:
        return L

    mid = len(L) // 2
    left = MergeSort(L[:mid])
    right = MergeSort(L[mid:])

    return merge(left, right)

def merge(left: List[int], right: List[int]) -> List[int]:
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result
````

Наприклад: `[[1, 2], [3, 4]] -> [1 4]`

````python
#2 Альтернативний варіант
def get_diag(X):
    diag = []
    for i in range(min(X.shape)):
        diag.append(X[i, i])
    return np.array(diag)
````

Завдання 9: Написати функцію яка приймає на вхід двомірний масив, а повертає масив елементів які знаходяться на другорядній діагоналі. Приклад: `[[1, 2], [3, 4]] -> [2 3]`

````python
# 1
def sec_diag(X):
    return np.diagonal(np.fliplr(X))
````

````python
# 2
def sec_diag(X):
    rows, cols = X.shape
    diag = []
    for i in range(min(rows, cols)):
        diag.append(X[i, cols-1-i])
    return np.array(diag)
````

Завдання 10: Кого більше - чоловіків чи жінок?

````python
def male_or_female(dataset):
    gender_counts = dataset['Sex'].value_counts()
    m = gender_counts.get('male',0)
    f = gender_counts.get('female',0)
    print(f"Кількість чоловіків: {m}\nКількість жінок: {f}")
    if m > f:
        print("Чоловіків більше")
    else:
        print("Жінок більше")
    

male_or_female(dt)
````

Завдання 11: Виведіть наступні характеристики пасажирів: середній вік пасажирів, середній вік чоловіків, середній вік жінок, максимальний і мінімальний вік серед чоловіків і жінок.

````python
def avg_age(dt):
    result = dt["Age"].mean().round(2)
    print(f"Середній вік пасажирів: {result}")

def avg_age_by_sex(dt):
    result = dt.groupby('Sex')['Age'].mean().round(2)
    for sex, age in result.items():
        print(f"Середній вік для {sex}: {age}")

def max_age(dt):
    result = dt.groupby('Sex')['Age'].max()
    for sex, age in result.items():
        print(f"Максимальний вік для {sex}: {age}")

def min_age(dt):
    result = dt.groupby('Sex')['Age'].min()
    for sex, age in result.items():
        print(f"Мінімальний вік для {sex}: {age}")

avg_age(dt)
avg_age_by_sex(dt)
max_age(dt)
min_age(dt)
````

 ## Знайомство з бібліотекою matplotlib

Для початку ознайомтесь з [документацією](https://matplotlib.org/)

````python
from matplotlib import pylab as plt
%matplotlib inline
````

````python
def f1(x):
    return np.sin(x) / x

def f2(x):
    return np.cos(x) * (x**2 + 1)

def f3(x):
    return np.cos(x)**2 - (np.sin(x))**2

def f4(x):
    return np.cos(x)

def f5(x):
    return -np.sin(x) * (x**2 + 1)

# Генерація значень x
x = np.linspace(-5, 5, 100)

# Побудова окремих графіків
plt.figure(figsize=(10, 6))
plt.plot(x, f1(x), label='$f_1(x) = \\frac{\\sin(x)}{x}$')
plt.grid(True)
plt.legend()
plt.title('Графік $f_1(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x, f2(x), label='$f_2(x) = \\cos(x)(x^2 + 1)$')
plt.grid(True)
plt.legend()
plt.title('Графік $f_2(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x, f3(x), label='$f_3(x) = \\cos^2(x) - \\sin^2(x)$')
plt.grid(True)
plt.legend()
plt.title('Графік $f_3(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x, f4(x), label='$f_4(x) = \\cos(x)$')
plt.grid(True)
plt.legend()
plt.title('Графік $f_4(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x, f5(x), label='$f_5(x) = -\\sin(x)(x^2 + 1)$')
plt.grid(True)
plt.legend()
plt.title('Графік $f_5(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````
